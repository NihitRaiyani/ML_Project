```python
import ast

def _to_list(x):
    if not x: return []
    if isinstance(x, list): return x
    try: return ast.literal_eval(x)
    except: return [s.strip() for s in str(x).split(',') if s.strip()]

def convert3(cast_cell, crew_cell):
    # get raw items
    cast_items = _to_list(cast_cell)
    crew_items = _to_list(crew_cell)
    names = []
    # helper to extract a name from an item (dict or string)
    def name_of(e):
        if isinstance(e, dict):
            return e.get('name') or e.get('original_name') or e.get('character') or e.get('actor')
        return str(e)
    # take from cast first
    for e in cast_items:
        n = name_of(e)
        if n: names.append(n)
        if len(names) == 3: return names
    # if not enough, prefer crew members in Acting department or actor-like jobs
    for e in crew_items:
        if isinstance(e, dict) and (e.get('department','').lower()=='acting' or e.get('job','').lower() in ('actor','cast','performer')):
            n = name_of(e)
            if n and n not in names:
                names.append(n)
        if len(names) == 3: return names
    # fallback: add any remaining crew names
    for e in crew_items:
        n = name_of(e)
        if n and n not in names:
            names.append(n)
        if len(names) == 3: break
    return names

# apply to dataframe (adjust column names if yours differ)
movies['cast'] = movies.apply(lambda r: convert3(r['cast'], r.get('crew','')), axis=1)
print(movies[['title','cast']].head(10).to_string())
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new = movies[['movie_id','title','tags']]
new['tags'] = new['tags'].apply(lambda x:" ".join(x))
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x:x.lower())
new.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)




```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words='english',max_features=5000)
```

```python
vectors = cv.fit_transform(new['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
